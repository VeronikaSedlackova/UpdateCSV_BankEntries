```python
def addIfSameBankNameInDict(row, listLines, writerForDict):
    sameLetter = ""
   
    for rowInOld in listLines: 
        
        # add IDP URL 
        if idpURL:                                    
            for idp in idpURL:
                if idp in row[1] and ("Sparkasse" in row[1] or "sparkasse" in row[1]):
                    row[6] = idpURL[idp] + row[5] + "/.well-known"
                elif idp in row[1]:
                    row[6] = idpURL[idp]

        # already filled out  
        if row[3] and row[4]:
            writerForDict.writerow(row)
            return "true"
            
        elif row[1][0] == rowInOld[1][0]:
            sameLetter = "true"
            
        # add productive URL and adapterID from dictionary
        if sameLetter == "true" and row[1][0] != rowInOld[1][0]:  
            for key in bankURLs:
                if key in row[1]:
                    if "Sparkasse" == key or "sparkasse" == key:  # Exception for sparkasse (bankcode in URL)
                        row[3] = bankURLs[key] + row[5]           # add URL
                    else:    
                        row[3] = bankURLs[key]                    # add URL
                                        
                    for value in bankAdapterId:
                        if value in row[1]:   
                            row[4] = bankAdapterId[value]         # add AdapterID
                            writerForDict.writerow(row)
                            return "true" 
            
            return "false"     
        else:
            continue
                            
```

```python
with open('addedData.csv', "r", encoding="utf-8") as file:
    readCSV = csv.reader(file)
      
    # resulting file 
    addDataFromDict = open('addedData_final.csv', 'w')
    writerForDict = csv.writer(addDataFromDict)
            
    # previous CSV file from GitLab
    inFile = open('aspsp-adapter-config_Sorted.csv','r')             
    readerInFile = csv.reader(inFile)
    
    
    ## filters individual names from previous file
    listLinesForSameBankName = []
    counter = 0

    for line in readerInFile:
        if line[1] in str(listLinesForSameBankName):
            continue
        else:
            listLinesForSameBankName += [line]  
            counter = counter + 1
            
    inFile.close()
       
    
    ## look up names in dictionary and add if equal  
    counterForNotFound = 0
    counterForEqualBanks = 0
        
    for line in readCSV:
        booleanForSuccess = addIfSameBankNameInDict(line, listLinesForSameBankName, writerForDict)
        
        if booleanForSuccess == "false":
            counterForNotFound = counterForNotFound + 1
            writerForDict.writerow(line)
        else:
            counterForEqualBanks = counterForEqualBanks + 1
        
        
    print("Data could not be added to", counterForNotFound, "row(s)")
    print("Add successfully: ", counterForEqualBanks, "row(s)")
```
